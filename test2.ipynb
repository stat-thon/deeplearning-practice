```python
import torch
import torchvision
import torchvision.transforms.v2 as v2
import numpy as np
import matplotlib.pyplot as plt

data = torchvision.datasets.CIFAR10(root = 'data/CIFAR10', train = True, download = True)
```

```python
from PIL import Image

image = Image.open('D:/repository/deeplearning-practice/deeplearning-practice/data/dog.png')
```

### Image Resizing

- v2.Resize
- v2.ScaleJitter
- v2.RandomShortestSize
- v2.RandomResize

```python
# Resize 
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title(f'Original image: {image.size}')

resized_img = v2.Resize((128, 128))(image) #   

plt.subplot(1, 2, 2)
plt.imshow(resized_img)
plt.title(f'Resized image: {resized_img.size}')

plt.show()
```

```python
# ScaleJitter 
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title(f'Original image: {image.size}')

jitterd_img = v2.ScaleJitter(target_size = (64, 64), scale_range = (1.0, 2.0))(image) #   

plt.subplot(1, 2, 2)
plt.imshow(jitterd_img)
plt.title(f'Scale jitterd image: {jitterd_img.size}')

plt.show()
```

```python
# RandomShortestSize 
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original image')

random_shortest_img = v2.RandomShortestSize(min_size = 10, max_size = 50)(image) # 

plt.subplot(1, 2, 2)
plt.imshow(random_shortest_img)
plt.title(f'Random image: {random_shortest_img.size}')

plt.show()
```

```python
# RandomReSize 
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title(f'Original image: {image.size}')

random_resize_img = v2.RandomResize(min_size = 30, max_size = 256)(image) # 

plt.subplot(1, 2, 2)
plt.imshow(random_resize_img)
plt.title(f'Random Resize image: {random_resize_img.size}')

plt.show()
```
